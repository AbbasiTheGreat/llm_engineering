```python
import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
load_dotenv(override=True)

GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
google_api_key = os.getenv("GOOGLE_API_KEY")
gemini = OpenAI( base_url = GEMINI_BASE_URL , api_key = google_api_key)
```

```python
def generate_email_subject(email: str):
    system_prompt = """You are an AI Email Subject Generator.

Your task is to analyze the content of an email and generate a concise, professional, and relevant subject line.

Guidelines:
- Understand the main purpose and intent of the email.
- Produce only one subject line.
- Keep the subject between 4 and 10 words whenever possible.
- Make the subject clear, specific, and actionable.
- Preserve important context such as project names, ticket numbers, dates, or customer names if they are central to the email.
- Do not invent information that is not present in the email.
- Avoid generic subjects such as "Update", "Hello", or "Regarding".
- Do not include prefixes like "Subject:", quotation marks, or explanations.
- Return only the generated subject line and nothing else.
"""

    user_prompt = f"""Generate an appropriate email subject for the following email.

Email:
{email}
"""

    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt},
    ]

    response = gemini.chat.completions.create(
        model="gemini-2.5-flash",
        messages=messages,
    )

    subject = response.choices[0].message.content.strip()
    
    return subject
```

```python
#example input for email
email = """Hi Sarah,

I hope you're doing well.

I would like to schedule a meeting sometime this week to discuss the updated timeline for the website redesign project. After reviewing the current progress and upcoming milestones, I believe it would be helpful to align on priorities, clarify any outstanding requirements, and address potential risks that could impact delivery.

If possible, could you let me know your availability on Wednesday or Thursday afternoon? If those times don't work, I'm happy to adjust to a time that's more convenient for you.

I look forward to our discussion.

Best regards, 
"""

```

```python
generate_email_subject(email)
```
